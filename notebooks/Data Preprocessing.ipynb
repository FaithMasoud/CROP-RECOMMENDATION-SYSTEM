```python
# PHASE 4 — DATA PREPROCESSING 

import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split

# -------------------------------------------
# STEP 1: Load the dataset
# Find which path works for your setup
# -------------------------------------------
import os
print("Current directory:", os.getcwd())

# Try both paths — use whichever works
try:
    df = pd.read_csv('../data/crop_data.csv')
    print("Loaded using ../data/crop_data.csv")
except FileNotFoundError:
    df = pd.read_csv('data/crop_data.csv')
    print("Loaded using data/crop_data.csv")

print("Dataset shape:", df.shape)

# -------------------------------------------
# STEP 2: Check for missing values
# -------------------------------------------
print("\n=== MISSING VALUES ===")
print(df.isnull().sum().sum(), "missing values found")

# -------------------------------------------
# STEP 3: Check for duplicates
# -------------------------------------------
print("\n=== DUPLICATES ===")
print(df.duplicated().sum(), "duplicate rows found")

# -------------------------------------------
# STEP 4: Separate Features (X) and Label (y)
# X = all columns except 'label'
# y = only the 'label' column
# -------------------------------------------
X = df.drop('label', axis=1)   # 7 input features
y = df['label']                 # crop name to predict

print("\n=== FEATURES (X) ===")
print("Shape:", X.shape)
print("Columns:", X.columns.tolist())

print("\n=== TARGET (y) ===")
print("Shape:", y.shape)
print("First 5 values:", y.head().tolist())

# -------------------------------------------
# STEP 5: Split into Training and Testing sets
# 80% training, 20% testing
# -------------------------------------------
X_train, X_test, y_train, y_test = train_test_split(
    X,               # input features
    y,               # target labels
    test_size=0.2,   # 20% for testing
    random_state=42, # reproducible split
    stratify=y       # balanced crop distribution
)

print("\n=== TRAIN/TEST SPLIT ===")
print(f"Total samples    : {len(X)}")
print(f"Training samples : {len(X_train)} (80%)")
print(f"Testing samples  : {len(X_test)} (20%)")

# -------------------------------------------
# STEP 6: Verify balanced crop distribution
# -------------------------------------------
print("\n=== CROP DISTRIBUTION IN TRAINING SET ===")
print(y_train.value_counts().sort_index())

print("\n=== CROP DISTRIBUTION IN TESTING SET ===")
print(y_test.value_counts().sort_index())

# -------------------------------------------
# STEP 7: Final confirmation
# -------------------------------------------
print("\n" + "="*45)
print("✅ Phase 4 Preprocessing Complete!")
print(f"   X_train shape : {X_train.shape}")
print(f"   X_test shape  : {X_test.shape}")
print(f"   y_train shape : {y_train.shape}")
print(f"   y_test shape  : {y_test.shape}")
print("="*45)
```
